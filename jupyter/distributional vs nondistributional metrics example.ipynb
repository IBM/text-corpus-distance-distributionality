```python
# illustration of perturbation of points
# import plotting and experiment function
import sys
sys.path.insert(0, '../src')
sys.path.insert(0, '../../comparing-corpora/src')

from illustrate_distributionality import *

# illustative example in 2D.  
# Blue points are from f_0, and orange from f_1

# fully paired: each item from f_1 is jitted a bit from one in f_0
X0, X1 = gen_data(draw=True, pair_proportion=1.0)
# Randomly select some points from the f_1 sample and perturb them (add large-standard deviation noise)
# so that they are further from their paired point in f_0 than otherwise
X1_p = perturb_points(X=X1, ppoints=0.1, perturb_std = 3.0, draw=True)

# now not paired, just an independent draw from the same sample (cluster centroids with specified standard deviation)
X0, X1 = gen_data(draw=True, pair_proportion=0.0)
# Randomly select some points from the f_1 sample and perturb them (add large-standard deviation noise)
# so that they are further from their paired point in f_0 than otherwise
X1_p = perturb_points(X=X1, ppoints=0.1, perturb_std = 3.0, draw=True)

# perturb points by a given radius rather than white noise (perturbation should be distance away if aspect ratio is 1)
X1_p = perturb_points(X=X1, ppoints=0.1, perturb_std = 0.0, perturb_radius=5, draw=True)

X1_p = perturb_points(X=X1, ppoints=0.99, perturb_std = 0.0, perturb_radius=10, draw=True)

# now do a combination of pairing and not, X1 should look more like X0 but not completely
X0, X1 = gen_data(draw=True, pair_proportion=0.8)

# move a group of points together in a given direction
X0, X1 = gen_data(draw=False, pair_proportion=0.0)

# force it to be away from the centroid
X1_p = X0.perturb_grouped_points(X1=X1, perturb_radius=2.0, draw=True, force_away=True)
```

```python
# show larger result of changing the perturbation sample size and standard deviation, for 2-D

n_samples = 300 # sample size of each draw
nrepetitions = 20 # repetitions of sampling/perturbation process process
# three different fixed perturbation sample sizes
nvec = np.maximum(n_samples * np.array([0.05, 0.1, 0.2, 0.5, 0.75, 1.0]), 1).round().astype(int)

# define alternative versions of DC and PR distance with different k-nearest neighbors
# rather than the default of 5

DC2 = lambda corpus1, corpus2: dc_distance(corpus1, corpus2, model=None, nearest_k=2)
PR2 = lambda corpus1, corpus2: pr_distance(corpus1, corpus2, model=None, nearest_k=2)
DC5 = lambda corpus1, corpus2: dc_distance(corpus1, corpus2, model=None, nearest_k=5)
PR5 = lambda corpus1, corpus2: pr_distance(corpus1, corpus2, model=None, nearest_k=5)
DC15 = lambda corpus1, corpus2: dc_distance(corpus1, corpus2, model=None, nearest_k=2)
PR15 = lambda corpus1, corpus2: pr_distance(corpus1, corpus2, model=None, nearest_k=2)

metrics = [classifier_distance, fid_distance, mauve_distance, PR2, PR5, PR15, DC2, DC5, DC15, IRPR_distance]#, IRPR_euc_distance]
metric_names = ['CLASSIFIER', 'FID', 'MAUVE', 'PR_2', 'PR_5', 'PR_15', 'DC_2', 'DC_5', 'DC_15', 'IRPR']#, 'IRPR_EUC']
```

```python
# compare the metrics on a spectrum of pairing

n_features = [2,6]
res = []
pair_ranges = np.linspace(start=0, stop=1, num=21).round(1).tolist()

for nf in n_features:
    # have to do in loop, there is some issue defining lambda functions in list/dict comprehension without overwriting
    gen_func = lambda pp: gen_data(n_features=nf, n_samples=n_samples, draw=False, pair_proportion=pp, jitter_std=0.0)
    res.append(run_distributional_comparison_pairing(metrics=metrics, metric_names=metric_names, gen_func=gen_func,
                                             repetitions=nrepetitions, pair_proportions=pair_ranges))
for nf, rr in zip(n_features, res):
    print('Pairing proportion spectrum, {} features'.format(nf))
    plot_distributional_comparison(rr, boxplot=True, decimals=1, ncolumns=5)
```

```python
# now generate with different numbers of features and fixed perturbation parameters

# can't do 1-dimensional since some of the metrics don't accept non-matrices
nfeatures_vec = np.arange(start=2, stop=7).astype(int)
gen_func = lambda nf: gen_data(n_features=nf, n_samples=n_samples, draw=False, pair_proportion=0.0)

# note, need to define with defaults, and not lambda function, because the parameter defaults need
# to be accessed

# no perturbation (for baseline)
def no_perturb_func(X, npoints=0, perturb_std=0.0, perturb_radius=0):
    return perturb_points(X, npoints=npoints, perturb_std=perturb_std, draw=False)
# perturb 5% of points
def perturb_sd_func1(X, npoints=nvec[0], perturb_std=3.0, perturb_radius=0):
    return perturb_points(X, npoints=npoints, perturb_std=perturb_std, draw=False)
# perturb 10% of points
def perturb_sd_func2(X, npoints=nvec[1], perturb_std=3.0, perturb_radius=0):
    return perturb_points(X, npoints=npoints, perturb_std=perturb_std, draw=False)
# perturb 20% of points
def perturb_sd_func3(X, npoints=nvec[2], perturb_std=3.0, perturb_radius=0):
    return perturb_points(X, npoints=npoints, perturb_std=perturb_std, draw=False)
# perturb 5% of points
def perturb_rad_func1(X, npoints=nvec[0], perturb_radius=3.0, perturb_std=0):
    return perturb_points(X, npoints=npoints, perturb_std=perturb_std, draw=False)
# perturb 10% of points
def perturb_rad_func2(X, npoints=nvec[1], perturb_radius=3.0, perturb_std=0):
    return perturb_points(X, npoints=npoints, perturb_std=perturb_std, draw=False)
# perturb 20% of points
def perturb_rad_func3(X, npoints=nvec[2], perturb_radius=3.0, perturb_std=0):
    return perturb_points(X, npoints=npoints, perturb_std=0, perturb_radius=perturb_radius, draw=False)
pf_std = [perturb_sd_func1, perturb_sd_func2, perturb_sd_func3]
pf_rad = [perturb_rad_func1, perturb_rad_func2, perturb_rad_func3]


# no perturbation
res_np = run_distributional_comparison_nfeatures(metrics=metrics, metric_names=metric_names,
                                              gen_func=gen_func, perturb_func=no_perturb_func,
                                              repetitions=nrepetitions, nfeatures_vec=nfeatures_vec)

# some perturbation of standard deviation
res_p = [run_distributional_comparison_nfeatures(metrics=metrics, metric_names=metric_names,
                                                 gen_func=gen_func, perturb_func=ff,
                                                 repetitions=nrepetitions, nfeatures_vec=nfeatures_vec) for ff in pf_std]

# by radius
res_r = [run_distributional_comparison_nfeatures(metrics=metrics, metric_names=metric_names,
                                                 gen_func=gen_func, perturb_func=ff,
                                                 repetitions=nrepetitions, nfeatures_vec=nfeatures_vec) for ff in pf_rad]

print('varying features from {}--{}, no perturbation'.format(nfeatures_vec[0], nfeatures_vec[-1]))
plot_distributional_comparison(res_np, boxplot=True)

for nn, rr in zip(nvec, res_p):
    print('varying features from {}--{}, with fixed {} points perturbation by standard deviation'.format(nfeatures_vec[0], nfeatures_vec[-1], nn))
    plot_distributional_comparison(rr, boxplot=True)
    
for nn, rr in zip(nvec, res_r):
    print('varying features from {}--{}, with fixed {} points perturbation by radius'.format(nfeatures_vec[0], nfeatures_vec[-1], nn))
    plot_distributional_comparison(rr, boxplot=True)
```
